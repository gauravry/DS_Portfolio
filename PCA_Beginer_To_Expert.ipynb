```python
from sklearn.linear_model import LogisticRegression  # for Logistic Regression algorithm
from sklearn.model_selection import train_test_split  # to split the dataset for training and testing
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris  # to load the iris dataset
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA

# Load the iris dataset
iris = load_iris()
X = iris.data[:5]  # Selecting only the first 5 rows of iris.data
y = iris.target[:5]

# Step 1: Using PCA for Eigenvalues and Eigenvectors
pca = PCA()
pca.fit(X)

fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(X)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    s=40,
)

ax.set_title("First three PCA dimensions")
ax.set_xlabel("1st Eigenvector")
ax.xaxis.set_ticklabels([])
ax.set_ylabel("2nd Eigenvector")
ax.yaxis.set_ticklabels([])
ax.set_zlabel("3rd Eigenvector")
ax.zaxis.set_ticklabels([])

print("PCA Eigenvalues (explained variance):")
print(pca.explained_variance_)
print("\nPCA Eigenvectors (principal components):")
print(pca.components_)

plt.show()
```

### Now We will Calculate the same using Numpy.

```python
# Calculate the covariance matrix
cov_matrix = np.cov(X.T)

# Calculate eigenvalues and eigenvectors using NumPy's linalg
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

print("\nNumPy Eigenvalues:")
print(eigenvalues)
print("\nNumPy Eigenvectors:")
print(eigenvectors)
```
