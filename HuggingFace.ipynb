```python
from transformers import pipeline

# Create a text generation pipeline
text_generation_pipeline = pipeline("text-generation")

# Generate text based on a prompt
prompt = "Once upon a time"
generated_text = text_generation_pipeline(prompt, max_length=50, do_sample=True)

# Print the generated text
print(generated_text[0]['generated_text'])
```

```python
# Generate text based on a prompt
prompt = "love is good i like it"
generated_text = text_generation_pipeline(prompt, max_length=100, do_sample=True)

# Print the generated text
print(generated_text[0]['generated_text'])
```

```python
from transformers import AutoTokenizer

# Load the tokenizer
model_name = "gpt2"  # You can use any other model you prefer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Tokenize the prompt
prompt = "love is good i like it"
tokens = tokenizer.tokenize(prompt)

# Print the tokens
print(tokens)
```

Lets see token generation for each of the models.

```python
from transformers import (
    GPT2Tokenizer, GPT2Model,
    BertTokenizer, BertModel,
    RobertaTokenizer, RobertaModel,
    XLNetTokenizer, XLNetModel,
    DistilBertTokenizer, DistilBertModel,
    T5Tokenizer, T5Model,
    BartTokenizer, BartModel,
    ElectraTokenizer, ElectraModel,
    AlbertTokenizer, AlbertModel
)

# Models
model_names = [
    "gpt2",
    "bert-base-uncased",
    "roberta-base",
    "xlnet-base-cased",
    "distilbert-base-uncased",
    "t5-small",
    "facebook/bart-base",
    "google/electra-small-discriminator",
    "albert-base-v2"
]

# Tokenizers
tokenizers = {
    "gpt2": GPT2Tokenizer,
    "bert-base-uncased": BertTokenizer,
    "roberta-base": RobertaTokenizer,
    "xlnet-base-cased": XLNetTokenizer,
    "distilbert-base-uncased": DistilBertTokenizer,
    "t5-small": T5Tokenizer,
    "facebook/bart-base": BartTokenizer,
    "google/electra-small-discriminator": ElectraTokenizer,
    "albert-base-v2": AlbertTokenizer
}

# Models
models = {
    "gpt2": GPT2Model,
    "bert-base-uncased": BertModel,
    "roberta-base": RobertaModel,
    "xlnet-base-cased": XLNetModel,
    "distilbert-base-uncased": DistilBertModel,
    "t5-small": T5Model,
    "facebook/bart-base": BartModel,
    "google/electra-small-discriminator": ElectraModel,
    "albert-base-v2": AlbertModel
}

# Prompt
prompt = "Once upon a time"

# Loop through each model
for model_name in model_names:
    print(f"Model: {model_name}")

    # Load tokenizer
    tokenizer = tokenizers[model_name].from_pretrained(model_name)

    # Tokenize prompt
    tokens = tokenizer.tokenize(prompt)

    # Print tokens
    print("Tokens:", tokens)
    print()
```
